```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd 
```

```python
theoffice=pd.read_csv('datasets/office_episodes.csv')
```

```python
theoffice.head(5)
```

```python
colors = []

for index,row in theoffice.iterrows():
    if row['scaled_ratings'] < 0.25 :
        colors.append('red')
    elif row['scaled_ratings'] < 0.50 : 
        colors.append('orange')
    elif row['scaled_ratings'] < 0.75 : 
        colors.append('lightgreen')
    else: 
        colors.append('darkgreen')
        
print(colors[:10])

size = []

for index,row in theoffice.iterrows():
    if row['has_guests'] == True:
        size.append(250)
    else: 
        size.append(25)
```

```python
theoffice['colors']= colors 
theoffice['size']=size
theoffice.info()
```

```python
non_guest_df = theoffice[theoffice['has_guests']== False]
guest_df = theoffice[theoffice['has_guests']== True]
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure()
plt.scatter(non_guest_df['episode_number'],non_guest_df['viewership_mil'],c=non_guest_df['colors'],s=non_guest_df['size'])
plt.scatter(guest_df['episode_number'],guest_df['viewership_mil'],c=guest_df['colors'],s=guest_df['size'],marker ='*')
plt.xlabel('Episode Number')
plt.ylabel('Viewership (Millions)')
plt.title('Popularity, Quality, and Guest Appearances on the Office')
plt.show()
```
